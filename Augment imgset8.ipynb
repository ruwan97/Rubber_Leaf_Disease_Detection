```python
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import array_to_img, img_to_array, load_img
import os

catname = "Colletotrichum"
resourcePath = "/content/drive/MyDrive/Research/ResearchData/"+catname

try:
    os.mkdir("/content/drive/MyDrive/Research/ResearchData/Colletotrichum/aug_"+catname)
except:
    pass

imglist = os.listdir(resourcePath)

datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True,
        fill_mode='nearest')

for img in imglist:
    img_path = os.path.join(resourcePath,img)

    print(img_path)
    
    img = load_img(img_path)  # this is a PIL image
    x = img_to_array(img)  # this is a Numpy array with shape (3, 150, 150)
    x = x.reshape((1,) + x.shape)  # this is a Numpy array with shape (1, 3, 150, 150)

    # the .flow() command below generates batches of randomly transformed images
    # and saves the results to the `preview/` directory
    i = 0
    for batch in datagen.flow(x, batch_size=1,save_to_dir="/content/drive/MyDrive/Research/ResearchData/Colletotrichum/aug_"+catname, save_prefix='img', save_format='jpeg'):
        
        print(img_path + " - " + str(i))
        
        i += 1
        if i > 12:
            break  # otherwise the generator would loop indefinitely
```
